The file has all the fields wrapped in quotes, which means pandas is treating 
the entire line as a single string, even though the delimiter is a comma.

```python
# Using quotechar='"' to handle the quoted fields
data = pd.read_csv(file_path, header=0, delimiter=',', quotechar="'")
data.head()
```

```python
data.columns
```

Still there are quotation marks at the beginning and end of each row. Let's 
manually preprocess the file by stripping quotes and splitting on commas

```python
cleaned_lines = []

# open the csv file in read mode with UTF-8 encoding
with open(file_path, "r", encoding='utf-8') as f:
    for line in f:
        # strip outer quotes and new line, then split by commas
        line = line.strip().strip('"')
        cleaned_lines.append(line.split(',')) 

columns = cleaned_lines[0]
rows = cleaned_lines[1:]

data = pd.DataFrame(rows, columns=columns)
data.tail()
```

```python
data.shape
```

```python
data.columns
```

We don't need all of these columns. We would just keep the columns for region, 
sub-region, and country/area. 

```python
# dropping the undesired columns and keeping only the desired ones.
cols_to_drop = [
    'Global Code', 'Global Name', 'Region Code', 'Sub-region Code', 
    'Intermediate Region Code', 'Intermediate Region Name', 'M49 Code',
    'ISO-alpha2 Code', 'ISO-alpha3 Code', 'Least Developed Countries (LDC)',
    'Land Locked Developing Countries (LLDC)', 'Small Island Developing States (SIDS)'
    ]
data_1 = data.drop(cols_to_drop, axis = 1)
data_1.head()
```

```python
# renaming the column names
data_renamed = data_1.rename(
    columns = {
        'Region Name': 'region',
        'Sub-region Name': 'sub_region',
        'Country or Area': 'area'
    }
)
data_renamed.head()
```

### Adding columns for special groups

```python
# adding indicator columns for special groups and setting all values as 0
data_renamed['european_union_country'] = 0
data_renamed['least_developed_country'] = 0
data_renamed['land_locked_developing_country'] = 0
data_renamed['small_island_developing_state'] = 0
data_renamed['low_income_food_deficit_country'] = 0
data_renamed['net_food_importing_developing_country'] = 0
data_renamed.head()
```

```python
# Loading csv files containing names of countries distributed by each of the 
# six special groups and adding the countries to a list corresponding to each 
# of the 6 special groups.
eu_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_european_union.csv')
eu_list = list(eu_data['Area'].unique())

ldc_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_LDC.csv')
ldc_list = list(ldc_data['Area'].unique())

lifdc_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_LIFDC.csv')
lifdc_list = list(lifdc_data['Area'].unique())

lldc_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_LLDC.csv')
lldc_list = list(lldc_data['Area'].unique())

nfidc_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_NFIDC.csv')
nfidc_list = list(nfidc_data['Area'].unique())

sids_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/FAOSTAT_SIDS.csv')
sids_list = list(sids_data['Area'].unique())
```

```python
# updating the values of the 6 special group columns only where the country/area 
# is in the corresponding list for that special group
data_renamed.loc[data_renamed['area'].isin(eu_list), 'european_union_country'] = 1
data_renamed.loc[data_renamed['area'].isin(ldc_list), 'least_developed_country'] = 1
data_renamed.loc[data_renamed['area'].isin(lifdc_list), 'low_income_food_deficit_country'] = 1
data_renamed.loc[data_renamed['area'].isin(lldc_list), 'land_locked_developing_country'] = 1
data_renamed.loc[data_renamed['area'].isin(nfidc_list), 'net_food_importing_developing_country'] = 1
data_renamed.loc[data_renamed['area'].isin(sids_list), 'small_island_developing_state'] = 1

data_renamed.head()
```

```python
# exporting the table containing indicator columns as csv file
data_renamed.to_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/indicator_columns.csv', index = False)
```
