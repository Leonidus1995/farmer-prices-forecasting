```python
import pandas as pd
import numpy as np
```

```python
# Loading raw data file
raw_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/Datasets_MS_Project/Trade_CropsLivestockIndicators_E_All_Data_(Normalized)/Trade_CropsLivestockIndicators_E_All_Data_(Normalized).csv')
raw_data.head()
```

```python
raw_data.shape
```

```python
raw_data.info()
```

```python
# removing irrelevant columns
columns_to_remove = ['Area Code (M49)', 'Item Code (CPC)', 'Indicator Code', 'Unit', 'Flag', 'Note']
filtered_data_1 = raw_data.drop(columns_to_remove, axis = 1)
filtered_data_1.head()
```

```python
filtered_data_1['Indicator'].unique()
```

```python
filtered_data_1['Indicator'].value_counts()
```

```python
# list of desired indicators
indicators_to_keep = [
       'Import dependency ratio', 
       'Self-sufficiency ratio',
       'Import market concentration index',
       'Export market concentration index',
       'Revealed comparative advantage index', 
       'Terms of trade'
       ]
```

```python
# filtering rows for desired indicators
filtered_data_2 = filtered_data_1.loc[filtered_data_1['Indicator'].isin(indicators_to_keep)]
```

```python
filtered_data_2['Year'].unique()
```

```python
# restructuring data from long to wide format
pivoted_data = filtered_data_2.pivot_table(
    index = ['Area Code', 'Area', 'Item Code', 'Item', 'Year Code', 'Year'],
    columns = 'Indicator',
    values = 'Value'
)
pivoted_data.head()
```

```python
# resetting row index
pivoted_data = pivoted_data.reset_index()
pivoted_data.head()
```

```python
# setting column index to None
pivoted_data.columns.name = None
pivoted_data.head()
```

```python
# total number of missing values for each column
pivoted_data.isna().sum()
```

```python
# renaming columns
cleaned_data_1 = pivoted_data.rename(
    columns = {
        'Area Code': 'area_code',
        'Area': 'area',
        'Item Code': 'item_code',
        'Item': 'item',
        'Year Code': 'year_code',
        'Year': 'year',
        'Import dependency ratio': 'import_dependency_ratio', 
        'Self-sufficiency ratio': 'self_sufficiency_ratio',
        'Import market concentration index': 'import_market_concentration_index',
        'Export market concentration index': 'export_market_concentration_index',
        'Revealed comparative advantage index': 'revealed_comparative_advantage_index', 
        'Terms of trade': 'terms_of_trade'
    }
)
cleaned_data_1.head()
```

```python
# exporting cleaned data as csv file
cleaned_data_1.to_csv('cleaned_datasets/trade_indicators_cleaned.csv', index=False)
```
