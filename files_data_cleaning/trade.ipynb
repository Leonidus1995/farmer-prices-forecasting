```python
# required libraries
import pandas as pd
import numpy as np
```

```python
# loading raw data file
raw_data = pd.read_csv('/Users/gurjitsingh/Desktop/MS Data Science/MS_Project_Python/Datasets_MS_Project/Trade_CropsLivestock_E_All_Data_(Normalized)/Trade_CropsLivestock_E_All_Data_(Normalized).csv')
raw_data.head()
```

```python
raw_data.info()
```

```python
# removing irrelevant columns
columns_to_remove = ['Area Code (M49)', 'Item Code (CPC)', 'Element Code', 'Unit', 'Flag', 'Note']
filtered_data_1 = raw_data.drop(columns_to_remove, axis=1)
filtered_data_1.head()
```

```python
filtered_data_1['Element'].unique()
```

```python
# filtering for desired elements and year
elements_to_keep = ['Import quantity', 'Export quantity', 'Export value', 'Import value']
filtered_data_2 = filtered_data_1.loc[(filtered_data_1['Year']>1990) & (filtered_data_1['Element'].isin(elements_to_keep))]
filtered_data_2.head()
```

```python
filtered_data_2['Element'].unique()
```

```python
filtered_data_2.shape
```

```python
len(filtered_data_2['Item'].unique())
```

```python
# restructuring data from long to wide format
pivoted_data = filtered_data_2.pivot_table(
    index = ['Area Code', 'Area', 'Item Code', 'Item', 'Year Code', 'Year'],
    columns = 'Element',
    values = 'Value'
)
pivoted_data.head()
```

```python
# resetting row index
pivoted_data = pivoted_data.reset_index()
pivoted_data.head()
```

```python
# setting column index to None
pivoted_data.columns.name = None
pivoted_data.head()
```

```python
len(pivoted_data['Area'].unique())
```

```python
# total number of missing values for each column
pivoted_data.isna().sum()
```

```python
# renaming columns
cleaned_data = pivoted_data.rename(
    columns = {
        'Area Code': 'area_code',
        'Area': 'area',
        'Item Code': 'item_code',
        'Item': 'item',
        'Year Code': 'year_code',
        'Year': 'year',
        'Export quantity': 'export_quantity',
        'Export value': 'export_value',
        'Import quantity': 'import_quantity',
        'Import value': 'import_value'
    }
)
cleaned_data.head()
```

```python
# exporting cleaned data as csv file
cleaned_data.to_csv('cleaned_datasets/trade_cleaned.csv', index=False)
```
