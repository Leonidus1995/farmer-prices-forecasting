```python
# required packages
import pandas as pd
```

```python
# Loading data
raw_data = pd.read_csv('Datasets_MS_Project/Production_Crops_Livestock_E_All_Data_(Normalized)/Production_Crops_Livestock_E_All_Data_(Normalized).csv')
raw_data.head()
```

```python
raw_data.shape
```

```python
raw_data.info()
```

```python
# removing irrelevant columns
columns_to_remove = ['Area Code (M49)', 'Item Code (CPC)', 'Flag', 'Note', 'Element Code']
filtered_data_1 = raw_data.drop(columns_to_remove, axis=1)
filtered_data_1.tail()
```

```python
filtered_data_1['Element'].unique()
```

```python
len(filtered_data_1['Area'].unique())
```

```python
len(filtered_data_1['Item'].unique())
```

```python
# loading cleaned producer price data
pp_data = pd.read_csv('cleaned_datasets/producer_prices_cleaned.csv')
pp_data.head()
```

```python
pp_list = pp_data['Item'].unique().tolist()
cp_list = filtered_data_1['Item'].unique().tolist()

# Finding common items in the two datasets
common_elements = set(pp_list) & set(cp_list)
len(common_elements)
```

```python
common_elements
```

```python
filtered_data_1['Element'].unique()
```

```python
# number of missing values per column
filtered_data_1.isna().sum()
```

```python
filtered_data_1['Element'].value_counts()
```

```python
# filtering data to keep the data after 1990 and removing rows for 'Stocks' element
filtered_data_2 = filtered_data_1.loc[(filtered_data_1['Year']>1990) & (filtered_data_1['Element']!='Stocks')]
filtered_data_2.shape
```

```python
filtered_data_2.head()
```

```python
# restructuring data from long to wide format
filtered_data_2_pivoted = filtered_data_2.pivot_table(
    index=['Area Code','Area','Item Code','Item','Year','Year Code'],
    columns='Element',
    values='Value'
)
filtered_data_2_pivoted.head()
```

```python
filtered_data_2_pivoted = filtered_data_2_pivoted.reset_index()
filtered_data_2_pivoted.head()
```

```python
filtered_data_2_pivoted.columns.name = None
filtered_data_2_pivoted.head()
```

```python
filtered_data_2_pivoted.info()
```

```python
filtered_data_2_pivoted.columns
```

```python
# renaming columns
cleaned_data = filtered_data_2_pivoted.rename(
    columns = {
        'Area Code': 'area_code', 
        'Area': 'area', 
        'Item Code': 'item_code', 
        'Item': 'item', 
        'Year': 'year', 
        'Year Code': 'year_code',
        'Area harvested': 'area_harvested', 
        'Laying': 'laying', 
        'Milk Animals': 'milk_animals',
        'Producing Animals/Slaughtered': 'producing_animals_or_slaughtered', 
        'Production': 'production', 
        'Yield': 'yield',
        'Yield/Carcass Weight': 'yield_or_carcass_weight'
    }
)
cleaned_data.head()
```

```python
# exporting cleaned data as csv file
cleaned_data.to_csv('cleaned_datasets/crop_production_cleaned.csv', index=False)
```
