```python
raw_data = pd.read_csv('Datasets_MS_Project/Fertilizers/FertilizersNutrient_E_All_Data_(Normalized)/Inputs_FertilizersNutrient_E_All_Data_(Normalized).csv', encoding = 'latin1')
raw_data.head()
```

We want to keep all the 7 elements in the "Element" column as they all are important 
and provide valuable information. Moreover, we would have to make a separate 
dataset for combinations of each of the three nutrients in the "Item" column with 
the 7 elements. We dont't have information about the combined total fertilizers. 

## Nitrogen fertilizers

```python
# Filter the data for N nutrient
filtered_data_N = raw_data.loc[raw_data['Item']=='Nutrient nitrogen N (total)']
filtered_data_N.head()
```

```python
filtered_data_N.info()
```

```python
filtered_data_N['Element'].value_counts()
```

```python
# restructuring data from long-format to wide-format
pivoted_data_N = filtered_data_N.pivot_table(
    index = ['Area Code', 'Area', 'Year Code', 'Year'],
    columns = 'Element',
    values = 'Value'
)

# resetting row index
pivoted_data_N.reset_index(inplace=True)

# setting name of column index to "None"
pivoted_data_N.columns.name = None

pivoted_data_N.head()
```

```python
pivoted_data_N.info()
```

"Export Quantity" and "Production" columns have significant missing values. However, 
they can be imputed later on.

```python
# Renaming columns
cleaned_data_N = pivoted_data_N.rename(
    columns = {
        'Area Code': 'area_code',
        'Area': 'area',
        'Year Code': 'year_code',
        'Year': 'year',
        'Agricultural Use': 'N2_for_Ag_use',
        'Export Quantity': 'N2_export_quantity',
        'Import Quantity': 'N2_import_quantity',
        'Production': 'N2_production',
        'Use per area of cropland': 'N2_use_per_area_of_cropland',
        'Use per capita': 'N2_use_per_capita',
        'Use per value of agricultural production': 'N2_use_per_value_of_Ag_production'
    }
)
cleaned_data_N.head()
```

```python
# Exporting cleaned data as csv file
cleaned_data_N.to_csv('cleaned_datasets/N2_fertilizers_cleaned.csv', index=False)
```

There are significant missing values in the "Production" for K2O nutrient. It
